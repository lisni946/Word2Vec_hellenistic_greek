## Preprocessing

```python
# Tokenizing
import nltk
import cltk
import string
from nltk.corpus.reader import WordListCorpusReader
from cltk.tokenize.sentence import TokenizeSentence
tokenizer = TokenizeSentence('greek')
```

```python
# locate corpus filepath, followed by individual file names in the [].
reader = WordListCorpusReader('/Users/christinaoaks/cltk_data/originals/tlg', ['001.txt', '002.txt', '003.txt', '004.txt', '005.txt', '006.txt', '007.txt', '008.txt', '009.txt', '010.txt', '011.txt', '012.txt'])

# convert to string
s = str(reader.words())
```

```python
# Stops
from nltk.tokenize.punkt import PunktLanguageVars
from cltk.stop.greek.stops import STOPS_LIST

p = PunktLanguageVars()
tokens = p.word_tokenize(s.lower())
s2 = [w for w in tokens if not w in STOPS_LIST]
s3 = ' '.join(s2)
```

```python
# remove commas and quotation marks
remover = s3.replace("'", '').replace(",", '').replace("(", '').replace(")", '')
s4 = str(remover).lower()
```

```python
# finally, tokenize the sentences
greek_corpus_sentence_tokens = tokenizer.tokenize_sentences(s4)
```

```python
# nummber of tokenized sentences
print(len(greek_corpus_sentence_tokens))
```
